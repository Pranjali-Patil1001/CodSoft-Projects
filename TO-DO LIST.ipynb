```python
import tkinter as tk
from tkinter import ttk
from tkinter import messagebox
import sqlite3 as sql

tasks=[]
def AddTask():
    task_str = task_field.get()
    if len(task_str) == 0 :
        messagebox.showinfo('Task should contain minimum one charater!!')
    else:
        tasks.append(task_str)
        the_cursor.execute('Insert into tasks values(?)',(task_str,))
        Update_list()
        task_field.delete(0,'end')

def Update_list():
    clear_list()
    for task in tasks:
        task_listbox.insert('end',task)
        
def delete_task():
    try:
        value = task_listbox.get(task_listbox.curselection())
        if value in tasks:
            tasks.remove(value)
            Update_list()
            the_cursor.execute('Delete from tasks where title = ?',(value,))
    except:
        messagebox.showinfo('No task selected!! Please select a task.')
        
            
def delete_all():
    message_box = messagebox.askyesno('Are you sure to delete all tasks ?')
    if message_box == True:
        while(len(tasks)!= 0):
            tasks.pop()
        the_cursor.execute('Delete from tasks')
        Update_list()
        
def clear_list():
    task_listbox.delete(0,'end')
    
def close():
    print(tasks)
    guiWindow.destroy()
    
def retrive():
    while(len(tasks)!=0):
        tasks.pop()
    for r in the_cursor.execute('Select title from tasks'):
        tasks.append(r[0])

if __name__ == "__main__":
    guiWindow = tk.Tk()
    guiWindow.title("TO-DO LIST APPLICATION")
    guiWindow.geometry("600x600+850+350")
    guiWindow.resizable(0,0)
    guiWindow.configure(bg="#7FFF00")
    connection = sql.connect('listOfTasks.db')
    the_cursor = connection.cursor()

    the_cursor.execute('Create table if not exists tasks (title text)')
    tasks = []  

    header=tk.Frame(guiWindow,bg = "#7FFF00")
    functions_frame = tk.Frame(guiWindow,bg = "#7FFF00")
    listbox_frame = tk.Frame(guiWindow,bg = "#7FFF00")
    
    header.pack(fill="both")
    functions_frame.pack(side="left",expand=True,fill="both")
    listbox_frame.pack(side="right",expand = True,fill="both")
    
    header_label = ttk.Label(header, text = "THE TO-DO LIST", font = ("Helvetica", "30", "bold") , 
                             background = "#7FFF00", foreground = "#8B4513")
    header_label.pack(padx =20 , pady = 20)
    task_label = ttk.Label(functions_frame,text="Enter a Task :",font = ("Helvetica","21","bold"),
                           foreground = "#000000")
    task_label.place(x = 30, y = 40)
    task_field = ttk.Entry(functions_frame,font=("Consolas","14"),width =18, 
                           background = "#FFF8DC",foreground = "#A52A2A")
    task_field.place(x=30,y=80)


    add_button = ttk.Button (
    functions_frame,
    text="Add Task",
    width = 30,    
    command = AddTask

    )


    del_button = ttk.Button (functions_frame,text="Delete Task",width = 30, command = delete_task,)


    delete_all_button = ttk.Button (functions_frame,text="Delete All Tasks ",width = 30, 
                                    command = delete_all)


    exit_button = ttk.Button (functions_frame,text="Exit",width = 30, command = close)


    add_button.place(x =30,y=120)

    del_button.place(x =30,y=160)

    delete_all_button.place(x =30,y=200)

    exit_button.place(x =30,y=240)


    task_listbox = tk.Listbox(listbox_frame,width=15,height=13,selectmode ='SINGLE',font = ("Helvetica", "20", "bold") ,
                              background ="#FFFFFF",foreground="#000000",selectbackground="#CD853F",selectforeground="#FFFFFF")

    task_listbox.place(x=10,y=20)


    retrive()

    Update_list()

    guiWindow.mainloop()

    connection.commit()

    the_cursor.close()
```
